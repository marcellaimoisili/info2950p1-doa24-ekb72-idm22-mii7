## Introduction

This project is centered around three datasets; the 2018 FIFA World Cup ratings, the 2018 World Happiness Report by the WHO and a 2017 Countries' Statistical Indicators dataset from UNData. These datasets together show both the conditions that existed during the highly coveted soccer tournament, and how they manifested in terms of soccer performance.  

For this project, we were interested in looking at countries that participated in the 2018 FIFA World Cup to answer different questions about their performance. Considering the top rated teams and the fact that unlike in other leagues, teams are sponsored by their respective countries, this is a great opportunity to examine the relationship (or lack thereof) between country dynamics and team performance. The are two metrics we will be focusing on to represent team performance; Categorical Ranking (Stage of Elimination), and Points attained. The three main metrics we will use to represent the level of support possible from countries are GDP (Gross Domestic Product) per Capita, percentage GVA (Gross Value Added) of industrial production, and healthy life expectancy.  

## Hypothesis

Given the relationships we decided to investigate, we formed the following hypotheses:

1. **Country GDP per Capita and Team Performance** : The country's GDP will be strongly and positively correlated with how the team performs.

2. **Industrial Production and Team Performance** : A country with more producing power in industry will correlate positively with how the team performs.

3. **Health/Life Expectancy and Team Performance** : The country's life expectancy indicates good living conditions. We are also predicting a positive correlation with performance for this metric.

#### Filter Dataset and Prepare for Merge

```python
#making dataframe for filtering the countries
filtered_countries = pd.DataFrame(country_spec, columns = ['Country or region','GDP per capita', 'Healthy life expectancy'])
filtered_countries = filtered_countries.rename(columns={'Country or region':'Squad'})
```

```python
#checking to see what countries were not included in the merge
fifa_data = pd.merge(fifa, filtered_countries, on='Squad')
set1 = set()
for i in fifa_data["Squad"]:
    set1.add(i)
for x in fifa["Squad"]:
    if x not in set1:
        print("Naming descrepancy found! -->",x)
```

##### Note:
While filtering our FIFA data to merge country health information with the countries represented in our `country_spec` dataset, we discovered that "South Korea" and "England" were represented as “Korea Republic” and “United Kingdom” respectively. 

We decided that the health information and other country specifics of the United Kingdom would serve our purposes in representing the state of England. As for South Korea, Korea Republic is simply the name of the country's soccer team. For this reason, we changed their names to “South Korea” and “United Kingdom” respectively for the data merge. 

Although this mismatch was resolved by a simple name change, we believe that it is important to acknowledge the significance of the differences in naming. Because the naming in our `country_spec` dataset refers to broader geographical regions, we decided to abide by the names used in this dataset.

```python
#replacing the mismatched values so the dataset can be merged
fifa["Squad"].replace({"Korea Republic":"South Korea", "England":"United Kingdom"}, inplace=True)
fifa_data = pd.merge(fifa, filtered_countries, on='Squad')
```

#### Changing Ranks 1, 2, 3, 4, to "Top Four"

##### Note:
Considering the dataset does not have numerical ranking for all the teams, it was worth organizing all the groups by category i.e., the top 4 teams will be grouped into the same ranking. This would allow for group comparisons.

```python
#creating a new column with categorical rankings per country
categorical_ranks=fifa_data["Rk"]
fifa_data["Categorical_Rank"] = categorical_ranks
```

```python
#replacing values with categories
print("These are the unique values before regrouping:", fifa_data["Categorical_Rank"].unique())
fifa_data["Categorical_Rank"] = fifa_data["Categorical_Rank"].replace({'1': 'Top Four',
                                                                       '2': 'Top Four',
                                                                       '3': 'Top Four',
                                                                       '4': 'Top Four'})
print ("These are the unique values after regrouping:", fifa_data["Categorical_Rank"].unique())
```

#### Creating Column in Dataframe to Index Colors (for graph visualization, not for analysis)

```python
#creating a new column with categorical colors
categorical_colors=fifa_data["Categorical_Rank"]
fifa_data["Colors"] = categorical_colors
```

```python
#replacing Values with colors
print("These are the unique values before regrouping:", fifa_data["Colors"].unique())
fifa_data["Colors"] = fifa_data["Colors"].replace({'Top Four': 3,
                                                    'QF': 2,
                                                    'R16': 1,
                                                    'GR': 0})
print ("These are the unique values after regrouping:", fifa_data["Colors"].unique())
```

#### Setting the Order for the Rankings for Proper Sorting

```python
#setting Order of the Rankings
fifa_data['Categorical_Rank'] = pd.Series(fifa_data['Categorical_Rank'], dtype="category")
fifa_data['Categorical_Rank'] = fifa_data['Categorical_Rank'].cat.reorder_categories(['GR', 'R16', 'QF', 'Top Four'], ordered=True)

#sorting the Values
fifa_data=fifa_data.sort_values(by=['Categorical_Rank'])
order=['GR', 'R16', 'QF', 'Top Four']

#resetting th index for new order
fifa_data=pd.DataFrame(fifa_data)
rankings=range(0,32)
fifa_data["Index"] = rankings
fifa_data = fifa_data.set_index("Index")
```

```python
#showing teams by categorical rank
team_and_rank=pd.DataFrame(data={'Squad': fifa_data["Squad"], 'Categorical_Rank': fifa_data["Categorical_Rank"]})
display(team_and_rank)
```

#### Preparing the Datasets for Mean, Variance, and Standard Deviation

```python
#creating dataFrame for mean, variance, and standard Deviation by categorical rank for future use
#we did not use all these variables
grouped_mean_data=pd.DataFrame(fifa_data.groupby(['Categorical_Rank']).mean())
grouped_var_data=pd.DataFrame(fifa_data.groupby(['Categorical_Rank']).var())
grouped_std_data=pd.DataFrame(fifa_data.groupby(['Categorical_Rank']).std())
print ("\nCategorically Grouped Mean:")
display (grouped_mean_data)
print ("\nCategorically Grouped Variance:")
display (grouped_var_data)
print ("\nCategorically Grouped Standard Deviation:")
display (grouped_std_data)
```

#### Significance of GDP per Capita (2018)

```python
#create random permutation of y response
def permuted_y_1():
    return np.random.permutation(bottom_four_gdp[2].copy())
```

```python
y_permutations = []
x_values = np.array(bottom_four_gdp[3])
rsq_perm = np.zeros(1000)

#calculate 1000 random permutation correlations and put them into an array
for i in range(len(rsq_perm)):
    y_permutations = permuted_y_1()
    perm_reg = LinearRegression().fit([x_values], [y_permutations])
    rsq = stats.linregress(x_values, y_permutations) 
    rsq_perm[i] = rsq[2]

#plotting the data
plt.hist(rsq_perm, bins = 30)
plt.title("Frequency of R-Squared values for Random Permutations\n(Points vs. GDP per Capita)", size=16, fontweight="bold")
plt.xlabel("R-Squared Values", size=12, fontweight="bold")
plt.ylabel("Occurences", size=12, fontweight="bold")
std_list=[]
for x in range (-3, 4):
    std_x = rsq_perm.mean()+(x*rsq_perm.std())
    std_list.append(std_x)
    plt.axvline(x=std_x, color="red", ls=":")

#adding the vertical lines to notate standard deviation and the where our r-squared value lies  
numbers=[-3, -2, -1, 0, 1, 2, 3]
for (x, num) in zip(std_list, numbers):
                plt.text (x, 15, s=str(abs(num))+" STDs", size = 12,
                          fontweight="bold", rotation=90, horizontalalignment='right', color="black")

plt.axvline(regression1[2], color="green", ls=":")
plt.show()

#printing the values
print ("Mean R-Squared using random order:", rsq_perm.mean())
print ("Standard Deviation of random order R-Squared Values:", rsq_perm.std())
```

#### Plot #14 Observation
This low mean value of r-squared amongst random permutations makes us confident that the correlation of 0.6583 (the green dotted line) between GDP per Capita and FIFA World Cup 2018 Performance holds statistical significance. The fact that our attained r-squared occurs is in the third standard deviation tells us that it is an uncommon value in random permutations. This tells us that the rank amongst the countries is statistically important.

## Explanation

Having found a higher correlation coefficient in our second analysis, we wondered why the initial analysis confounded the pattern. We speculated that this could be because of the unpredictability of sports i.e., upsets. For example, Germany was knocked out of the 2018 World Cup in the group stage despite having won the competition in 2014. Another reason the pattern could've been confounded is the that the group stage is based on a team’s performance over the course of three games, while the Round of 16, the next stage in the competition is based on a single game. Given this, for a team to make it out of the group stage, they need to be consistently better than the competition. Later in the competition, upsets are more likely to happen since teams only need to win one game to advance to the next stage. There is more luck and unpredictability in these stages and the trend we see in welfare statistics seems to become less consistent as the tournament progresses. The further a group goes in the competition, it seems that our chosen measures have a reduced effect on their performance.
Supporting Article:
https://www.sbnation.com/soccer/2018/6/13/17458288/2018-fifa-world-cup-bracket-schedule-results-scores-format

A similar trend seems to emerge in the Gross Value Added by Industry, except this trend can be explained using different reasoning. In Western and traditionally soccer enthusiast countries such as England and Spain, talented players are typically found at a young age and recruited into academy systems that help develop their talent and ultimately help them become professional soccer players. However, in countries where industry and factory work are essentials in the economy, talented soccer players are much less likely to have an opportunity to develop their skills, because they are culturally expected to take up factory work. As a result, these young men are much more likely to spend their time during their athletic and physical prime doing heavy duty factory work rather than honing their skills as professional athletes. According to sportanddev.org, this phenomenon is called the “muscle drain”. This phenomenon means that developing countries supply industrialized countries with soccer talent.
